```python
import pandas as pd
import re
import numpy as np
```

```python
df = pd.read_parquet("train-00000-of-00001-e62b919b003d79c2.parquet")
df = df.rename(columns={"full_article": "text"})
df.head()
```

# Preprocessing Khmer Text

```python
def segment_sentences(text):
    sentences = re.split(r'(?<=[។៕៖?!])\s*', text)
    return [s.strip() for s in sentences if s.strip()]

df['sentences'] = df['text'].apply(segment_sentences)
df[['text', 'summary', 'sentences']].head()
```

```python
from khmernltk import word_tokenize

def segment_words(sentences):
    return [word_tokenize(sentence) for sentence in sentences]

df['word_tokens'] = df['sentences'].apply(segment_words)
df[['text', 'summary', 'word_tokens']].head()
```

# Extractive Method 1:TextRank Sentence Ranking

```python
import networkx as nx
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

```

```python
def textrank_summarize(sentences, top_n=3):
    if len(sentences) == 0:
        return ""

    vectorizer = TfidfVectorizer(
        tokenizer=word_tokenize,
        lowercase=False
    )

    tfidf_matrix = vectorizer.fit_transform(sentences)
    similarity_matrix = cosine_similarity(tfidf_matrix)

    nx_graph = nx.from_numpy_array(similarity_matrix)
    scores = nx.pagerank(nx_graph)

    ranked = sorted(((scores[i], s) for i, s in enumerate(sentences)), reverse=True)
    summary = ' '.join([s for _, s in ranked[:top_n]])
    return summary
```

```python
df['textrank_summary'] = df['sentences'].apply(textrank_summarize)
df[['summary', 'textrank_summary']].head()

```

# Extractive Method 2: TF-IDF (Graph-Based)

```python
def tfidf_summarize(sentences, top_n=3):
    if len(sentences) == 0:
        return ""

    vectorizer = TfidfVectorizer(
        tokenizer=word_tokenize,
        lowercase=False
    )

    tfidf_matrix = vectorizer.fit_transform(sentences)
    sentence_scores = tfidf_matrix.sum(axis=1).A1

    ranked = sorted(
        ((score, s) for score, s in zip(sentence_scores, sentences)),
        reverse=True
    )

    summary = ' '.join([s for _, s in ranked[:top_n]])
    return summary
```

```python
df['tfidf_summary'] = df['sentences'].apply(tfidf_summarize)
df[['summary', 'tfidf_summary']].head()
```

# Extractive Method 3: Clustering-Based Extraction

```python
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
```

```python
def clustering_summarize(sentences, num_clusters=3):
    # Safety checks
    if not sentences:
        return ""

    if len(sentences) <= 1:
        return sentences[0]

    vectorizer = TfidfVectorizer(
        tokenizer=word_tokenize,
        lowercase=False
    )

    tfidf_matrix = vectorizer.fit_transform(sentences)

    # Reduce clusters if needed
    n_clusters = min(num_clusters, tfidf_matrix.shape[0])

    # If all vectors are identical → fallback
    if tfidf_matrix.nnz == 0:
        return sentences[0]

    kmeans = KMeans(
        n_clusters=n_clusters,
        random_state=42,
        n_init=10
    )

    labels = kmeans.fit_predict(tfidf_matrix)

    summary_sentences = []

    for cluster_id in range(n_clusters):
        idxs = np.where(labels == cluster_id)[0]

        #  FIX: skip empty clusters
        if len(idxs) == 0:
            continue

        cluster_matrix = tfidf_matrix[idxs]
        scores = cluster_matrix.sum(axis=1).A1

        if len(scores) == 0:
            continue

        best_idx = idxs[int(np.argmax(scores))]
        summary_sentences.append(sentences[best_idx])

    # Fallback if nothing selected
    if not summary_sentences:
        scores = tfidf_matrix.sum(axis=1).A1
        return sentences[int(np.argmax(scores))]

    return " ".join(summary_sentences)

```

```python
df['clustering_summary'] = df['sentences'].apply(clustering_summarize)
df[['summary', 'clustering_summary']].head()

```

```python
from rouge import Rouge
rouge = Rouge()
```

```python
def tokenize_for_rouge(text):
    return " ".join(word_tokenize(text))
```

```python
def evaluate_rouge_khmer(reference, candidate):
    ref_tok = tokenize_for_rouge(reference)
    cand_tok = tokenize_for_rouge(candidate)
    scores = rouge.get_scores(cand_tok, ref_tok, avg=True)
    return scores
```

```python
df['textrank_rouge'] = df.apply(
    lambda r: evaluate_rouge_khmer(r['summary'], r['textrank_summary']),
    axis=1
)

df['tfidf_rouge'] = df.apply(
    lambda r: evaluate_rouge_khmer(r['summary'], r['tfidf_summary']),
    axis=1
)

df['clustering_rouge'] = df.apply(
    lambda r: evaluate_rouge_khmer(r['summary'], r['clustering_summary']),
    axis=1
)
```
